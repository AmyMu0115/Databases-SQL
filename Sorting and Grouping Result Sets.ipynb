## Eliminating Duplicates - DISTINCT clause
At times, a select statement result set can contain duplicate values. 

For example, in the author table example, the country column litsts the two-letter country code of the author's country. But some of the authors come from the same country, the results set contains duplicates. 


To eliminate duplicates, we use the keyword distinct


## Group by clause
But what if we wanted to also know how many authors come from the same country?

In this case, we add the "group by" clause to the select statement. 

The "group by" clause groups a result into subsets that has matching values for one or more columns
