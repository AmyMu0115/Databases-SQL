# Adding rows to a table
* Create the table (CREATE TABLE statement)
* Populated with table with data:
   * INSERT statement
       * a Data Manipulation Language (DML) statement used to read and modify data
       


# Using the INSERT Statenemt

The syntax of the INSERT statement looks like this:

```python
INSERT INTO [TableName]
<([ColumnName],...)>
VALUES([VALUE]...)
```

For example:

```python
INSERT INTO AUTHOR
(AUTHOR_ID.LASTNAMe,FIRSTNAME,EMAIL,CITY,COUNTRY)
VALUES
('A1','CHONG','RAUL','RFC@IBM.COM','TORONTO','CA')
```

It is important that the number of values provided in the values clause is equal to the number of column names specified in the column name list. This ensures that each column has a value.

# Inserting multiple rows
Tables do not need to be populated one row at a time, mutiple rpws can be inserted by specifying each row in the values clause. 

```python
INSERT INTO AUTHOR
(AUTHOR_ID.LASTNAMe,FIRSTNAME,EMAIL,CITY,COUNTRY)
VALUES
('A1','CHONG','RAUL','RFC@IBM.COM','TORONTO','CA')
```
